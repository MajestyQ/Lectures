```python
import matplotlib.pyplot as plt
import numpy as np
import time
%matplotlib notebook
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# some X and Y data
x = np.arange(50)
y = np.random.randn(50)

line, = ax.plot(x, y)

# draw and show it
ax.relim() 
ax.autoscale_view(True,True,True)
fig.canvas.draw()
plt.show(block=False)

x=0

while True:
    try:
        y[:-1] = y[1:]
        y[-1:] = np.random.randn(1)

        # set the new data
        line.set_ydata(y)
        
        ax.relim() 
        ax.autoscale_view(True,True,True)

        fig.canvas.draw()

        time.sleep(0.1)
    except KeyboardInterrupt:
        break
```
