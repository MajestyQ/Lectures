```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

#eine Zahl in Form eines Strings wird als Parameter genommen. Der String ist eine Kommazahl und wird zu einer
#float umgewandelt. Dazu muss aber zuerst das Komma zu einem Punkt geÃ¤ndert werden
def comma_to_float (valstr):
    return float (valstr.decode("utf-8").replace(',','.'))

#hier werden die Messdaten des Spektrometers geladen
#ohne Fenster ...
lamb_og, inten_og = np.loadtxt( 'C:/Users/Praktikum-Pc1/Documents/AA/Himmel.txt' , skiprows =17 , converters = { 0 :comma_to_float, 1 :comma_to_float}, comments = '>',unpack=True)
#... und mit Fenster
lamb_mg, inten_mg = np.loadtxt( 'C:/Users/Praktikum-Pc1/Documents/AA/Fenster.txt' , skiprows =17 , converters = { 0 :comma_to_float, 1 :comma_to_float}, comments = '>',unpack=True)

#Hier wird der Plot aufgesetzt:
#Als erstes wird die Messreihe ohne Fenster geplottet
plt.figure(figsize=(15,10))
plt.plot(lamb_og, inten_og, label = 'ohne Fenster' )
#dann die mit Fenster
plt.plot(lamb_mg, inten_mg, label = 'mit Fenster' )
#unser Bild bekommt nocht einen Titel, Achsenbeschriftugn und eine Legende
plt.title( 'Gemessenes Sonnenspektrum mit und ohne Fenster' )
plt.xlabel( 'Wellenlaenge / nm' )
plt.ylabel( 'Intensitaet / b.E.' )
plt.legend()
plt.grid() 
plt.ylim(( 0 , 60000 ))
plt.xlim(( 250 , 900))
#zum Schluss noch speichern
plt.savefig( "C:/Users/Praktikum-Pc1/Documents/AA/Himmel1" , format = "pdf" )
```

```python
#Hier wird die Absorbption von Glas berechnet
#dabei werden die Werte, welche Null sind Ã¼bersprungen
A =1- inten_mg[inten_og>0] / inten_og[inten_og>0]
#Hier wird der zweite Plot initialisiert
plt.plot(lamb_mg[inten_og>0], A) 
plt.title( 'Absorption von Glas' ) 
plt.xlabel( 'Wellenlaenge / nm' ) 
plt.ylabel( 'Absorption / b.E.' ) 
#bei uns musste der Bereich auf negative Zahlen erweitert werden, da die Messwerte bei ohne Fenster teilweise kleiner sind, als die mit Fenster
plt.ylim(( -.2 , 1.05 ))
plt.xlim(( 320 , 800 )) 
#datei speichern
plt.savefig( "C:/Users/Praktikum-Pc1/Documents/AA/Absorption_Glas.pdf" , format = "pdf" )
```

```python
wellenl=np.array([452.3,440.2,433.8,430.4,426.7])
fehler=np.array([1,1,1,1,1])
quantenz=np.arange(7,12)

plt.errorbar(quantenz,wellenl,fehler, fmt=".")
plt.xlabel('Quantenzahl')
plt.ylabel('Wellenlaenge/nm')
plt.title('1. Nebenserie des Na-Atoms')

from scipy.optimize import curve_fit
def fit_func(m,E_Ry,E_3p,D_d):
    return 1.2398E3/(E_Ry/(m-D_d)**2-E_3p)

para = [-13.6,-3,-0.02]
popt, pcov = curve_fit(fit_func, quantenz, wellenl, sigma=fehler ,p0=para)

print("E_Ry=",popt[0], ",Standardfehler=", np.sqrt(pcov[0][0]))
print("E_3p=",popt[1], ",Standardfehler=", np.sqrt(pcov[1][1]))
print("D_d=",popt[2], ",Standardfehler=", np.sqrt(pcov[2][2]))
```

```python
chi2_=np.sum((fit_func(quantenz,*popt)-wellenl)**2/fehler**2)
dof=len(quantenz)-3 #dof:degreesof freedom, Freheitsgrad
chi2_red=chi2_/dof
print("chi2=", chi2_)
print("chi2_red=",chi2_red)
```

```python
from scipy.stats import chi2
prob=round(1-chi2.cdf(chi2_,dof),2)*100
print("Wahrscheinlichkeit:", prob,"%")
```

```python
plt.errorbar(quantenz,wellenl,fehler, fmt=".")
plt.xlabel('Quantenzahl')
plt.ylabel('Wellenlaenge/nm')
plt.ylim((400,1000))
plt.title('1. Nebenserie des Na-Atoms')
x=np.linspace(2.8,12.2, 100)
plt.plot(x, fit_func(x,*popt))
plt.savefig("data/1_Nebenserie.pdf", format="pdf")
```
