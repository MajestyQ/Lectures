### 1. Auswertung des Sonnenspektrums

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def comma_to_float(valstr):
    return float(valstr.decode("utf-8").replace(',','.'))

lamb_og, inten_og = np.loadtxt('data/HimmelslichtJTHog.txt', skiprows=17,
                              converters= {0:comma_to_float, 1:comma_to_float},
                              comments='>',unpack=True)
lamb_mg, inten_mg = np.loadtxt('data/HimmelslichtJTHmg.txt', skiprows=17,
                              converters= {0:comma_to_float, 1:comma_to_float},
                              comments='>',unpack=True)
```

```python
plt.plot(lamb_og, inten_og, label='ohne Fenster')
plt.plot(lamb_mg, inten_mg, label='mit Fenster')
plt.legend()
plt.grid()
plt.ylim((0,60000))
plt.xlim((250,900))
```

Absorption von Glas berechnen

```python
A=1-inten_mg/inten_og

plt.plot(lamb_mg,A)
plt.ylim((0,1))
plt.xlim((320,800))
```

### 2. Auswertung des Natriumspektrums

```python
lamb_Na1, inten_Na1 = np.loadtxt('data/NaSpektrumJTHschwacheLinien.txt', skiprows=17,
                              converters= {0:comma_to_float, 1:comma_to_float},
                              comments='>',unpack=True)
lamb_Na2, inten_Na2 = np.loadtxt('data/NaSpektrumJTHstarkeLinien.txt', skiprows=17,
                              converters= {0:comma_to_float, 1:comma_to_float},
                              comments='>',unpack=True)
lamb_Na3, inten_Na3 = np.loadtxt('data/NaSpektrumJTHstarkeLinien2.txt', skiprows=17,
                              converters= {0:comma_to_float, 1:comma_to_float},
                              comments='>',unpack=True)

plt.figure(figsize=(20,5))
plt.subplot(131)
plt.plot(lamb_Na1, inten_Na1)
plt.title('Natriumspektrum geringer Intensitt')
plt.yscale('log')
plt.ylim((0,80000))
plt.xlim((300,540))

plt.subplot(132)
plt.plot(lamb_Na2, inten_Na2)
plt.title('Natriumspektrum hoher Intensitt')
plt.yscale('log')
plt.ylim((10,80000))
plt.xlim((300,850))

plt.subplot(133)
plt.plot(lamb_Na3, inten_Na3)
plt.title('Natriumspektrum hoher Intensitt(2)')
plt.yscale('log')
plt.ylim((10,80000))
plt.xlim((300,850))
```
