```python
%matplotlib notebook


import matplotlib.pyplot as plt
import numpy as np

def comma_to_float(valstr):
    return float(valstr.decode("utf-8").replace(',', '.'))

lamb_og, inten_og=np.loadtxt('data/Himmelslicht_ohne_Fenster.txt', skiprows=17,
                             converters= {0:comma_to_float, 1:comma_to_float},
                             comments='>', unpack=True)

lamb_mg, inten_mg=np.loadtxt('data/Himmelslicht_mit_Fenster.txt', skiprows=17,
                             converters= {0:comma_to_float, 1:comma_to_float},
                             comments='>', unpack=True)

#skal_faktor = inten_og.max()/inten_mg.max()
```

```python
plt.plot(lamb_og, inten_og, label="ohne Fenster")#normierung
plt.plot(lamb_mg, inten_mg, label="mit Fenster")#normierung

plt.title('Gemessenes Sonnenspektrum mit und ohne Fenster')
plt.xlabel(r'WellenlÃ¤nge in $nm$')
plt.ylabel('IntensitÃ¤t [b.E.]')
plt.legend()
plt.grid()
plt.ylim((0, 60000))
plt.xlim((250, 900))
plt.savefig("data/Himmelsspektrum.pdf", format="pdf")

#plt.show()
```

```python
A = 1-(inten_mg)/(inten_og)#skal_faktor*

plt.plot(lamb_mg, A)
plt.title('Absoption von Glas')
plt.xlabel('WellenlÃ¤ngen in $nm$')
plt.ylabel('Absoprtion [b.E.]')
plt.ylim((0, 1))
plt.xlim((320, 800))
plt.savefig("data/Absorption_Glas.pdf", format="pdf")

#plt.show()
```

```python
wellenl=np.array([819, 568, 498, 467, 450, 439, 433, 429, 426, 419])

fehler=np.array([2, 2, 1, 1, 1, 2, 1, 1, 1, 2,])


quantenz=np.arange(3,13)

plt.errorbar(quantenz, wellenl, fehler, fmt=".")
plt.title('1. Nebenserie des Na-Atoms')
plt.xlabel('Quantenzahl')
plt.ylabel('WellenlÃ¤nge [nm]')
x=np.linspace(2.8,12.2,100)
plt.ylim((400, 1000))
plt.xlim((2, 14))

from scipy.optimize import curve_fit
def fit_func(m, E_Ry, E_3p, D_d):
    return 1.2398E3/(E_Ry/(m-D_d)**2-E_3p)
para = [-13.6, -3, -0.02]
popt, pcov = curve_fit(fit_func, quantenz, wellenl, sigma=fehler, p0=para)
print("E_Ry=", popt[0], ", Standartfehler =", np.sqrt(pcov[0][0]))
print("E_3p=", popt[1], ", Standartfehler =", np.sqrt(pcov[1][1]))
print("D_d=", popt[2], ", Standartfehler =", np.sqrt(pcov[2][2]))

chi2_=np.sum((fit_func(quantenz, *popt)-wellenl)**2/fehler**2)
dof=len(quantenz)-3
chi2_red=chi2_/dof
print("chi2=", chi2_)
print("chi2_red=", chi2_red)

from scipy.stats import chi2
prob=round(1-chi2.cdf(chi2_,dof),2)*100
print("Wahrscheinlichkeit:", prob,"%")

plt.text(5, 900, 'E_Ry='+str(round(popt[0],2))+', Standartfehler ='+
        str(round(np.sqrt(pcov[0][0]),2)))
plt.text(5, 850, 'E_3p='+str(round(popt[1],2))+', Standartfehler ='+
        str(round(np.sqrt(pcov[1][1]),2)))
plt.text(5, 800, 'D_d='+str(round(popt[2],2))+', Standartfehler ='+
        str(round(np.sqrt(pcov[2][2]),2)))
plt.text(5, 750, 'chi2='+str(round(chi2_)))
plt.text(5, 700, 'chi2_red='+str(round(chi2_red)))
        
plt.plot(x, fit_func(x, *popt))
plt.savefig("data/1_Nebenserie.pdf", format="pdf")
plt.show()
```
