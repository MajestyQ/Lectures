```python
import matplotlib.pyplot as plt
import numpy as np

def comma_to_float(valstr):
    return float(valstr.decode("utf-8").replace(',','.'))

lamb_og, inten_og = np.loadtxt('data\penne/sonnenlicht.txt', skiprows=17, 
                                 converters = {0:comma_to_float, 1: comma_to_float},
                                 comments='>', unpack = True)


lamb_mg, inten_mg = np.loadtxt('data\penne/sonnemitglas.txt', skiprows=17, 
                                 converters = {0:comma_to_float, 1: comma_to_float},
                                 comments='>', unpack = True)

skal_factor = inten_og.max() / inten_mg.max()

#Skalierung

plt.plot(lamb_og, inten_og, label = 'ohne Fenster')
plt.plot(lamb_mg, inten_mg * skal_factor, label = 'mit Fenster')
plt.title('Gemessenes Sonnenspektrum mit und ohne Fenster, skaliert')
plt.xlabel('Wellenlänge / nm')
plt.ylabel('Intensität / b.E.')
plt.legend()
plt.grid()
plt.ylim((0,65000))
plt.xlim((250,900))
plt.savefig("figures/Sonne_skaliertFJ.pdf", format = "pdf")


```

```python
#Keine Skalierung
%matplotlib notebook
plt.plot(lamb_og, inten_og, label = 'ohne Fenster')
plt.plot(lamb_mg, inten_mg, label = 'mit Fenster')
plt.title('Gemessenes Sonnenspektrum mit und ohne Fenster')
plt.xlabel('Wellenlänge / nm')
plt.ylabel('Intensität / b.E.')
plt.legend()
plt.grid()
plt.ylim((0,65000))
plt.xlim((250,900))
plt.savefig("figures/Sonne_FJ.pdf", format = "pdf")
```

```python
A = 1 - inten_mg/inten_og

plt.plot(lamb_mg, A)
plt.title('Abb. 4: Absoption von Glas')
plt.xlabel('Wellenlänge')
plt.ylabel('Absorption / b.E')
plt.ylim((0,1))
plt.xlim((320,820))

plt.savefig("figures/FJAbsorbtion3.pdf", format = "pdf")
```

```python
#Geringe Intensität Natriumspektrum

lamb, inten = np.loadtxt('data/Natrium_schwach2.txt', skiprows=17,
                              converters= {0:comma_to_float, 1:comma_to_float},
                              comments='>', unpack=True)

plt.plot(lamb, inten)
plt.title('Abb. 6: Natriumspektrum geringer Intensität')
plt.xlabel('Wellenlänge / nm')
plt.ylabel('Intensität / b.E.')
plt.semilogy()
plt.ylim((0,600000))
plt.xlim((300,540))

plt.savefig("figures/Natriumspektrum geringe Intensität.pdf", format = "pdf")
```
