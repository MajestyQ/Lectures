```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def comma_to_float(valstr):
    return float(valstr.decode("utf-8").replace(',','.'))
```

```python
lamb_sun, inten_sun=np.loadtxt('data/2017-05-16/sun.txt', skiprows=17,converters= {0:comma_to_float, 1:comma_to_float},comments='>', unpack=True)
```

```python
lamb_sun_window, inten_sun_window=np.loadtxt('data/2017-05-16/sun_window.txt', skiprows=17,converters= {0:comma_to_float, 1:comma_to_float},comments='>', unpack=True)
```

```python
plt.plot(lamb_sun, inten_sun, label='ohne Fenster')
plt.plot(lamb_sun_window, inten_sun_window, label='mit Fenster')
plt.title('Gemessenes Sonnenspektrum mit und ohne Fenster')
plt.xlabel('Wellenlaenge / nm')
plt.ylabel('Intensitaet / b.E.')
plt.legend()
plt.grid()
plt.ylim((0,65536))
plt.xlim((250,900))
plt.savefig("data/2017-05-16/sun.pdf", format="pdf")
```

```python
A=1-inten_sun_window/inten_sun
```

```python
plt.plot(lamb_sun, A)
plt.title('Absorpton von Glas')
plt.xlabel('Wellenlaenge / nm')
plt.ylabel('Absorption / b.E.')
plt.ylim((0,1))
plt.xlim((320,800))
plt.savefig("data/2017-05-16/Absorption_Glas.pdf", format="pdf")
```

```python
plt.plot(lamb_sun, inten_sun)
plt.title('Sonnenspektrum')
plt.xlabel('Wellenlaenge / nm')
plt.ylabel('Intensitaet / b.E.')
plt.ylim((0,65536))
plt.xlim((350,800))
plt.savefig("data/2017-05-16/Fraunhofer.pdf", format="pdf")
```
