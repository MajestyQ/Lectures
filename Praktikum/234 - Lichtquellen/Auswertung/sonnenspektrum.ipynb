```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

#eine Zahl in Form eines Strings wird als Parameter genommen. Der String ist eine Kommazahl und wird zu einer
#float umgewandelt. Dazu muss aber zuerst das Komma zu einem Punkt geÃ¤ndert werden
def comma_to_float (valstr):
    return float (valstr.decode("utf-8").replace(',','.'))

#hier werden die Messdaten des Spektrometers geladen
#ohne Fenster ...
lamb_og, inten_og = np.loadtxt( 'F:/WinPython-64bit-3.4.4.4Qt5/notebooks/data/AnnaLeoni/ohne_fenster.txt' , skiprows =17 , converters = { 0 :comma_to_float, 1 :comma_to_float}, comments = '>',unpack=True)
#... und mit Fenster
lamb_mg, inten_mg = np.loadtxt( 'F:/WinPython-64bit-3.4.4.4Qt5/notebooks/data/AnnaLeoni/mit_fenster.txt' , skiprows =17 , converters = { 0 :comma_to_float, 1 :comma_to_float}, comments = '>',unpack=True)

#Hier wird der Plot aufgesetzt:
#Als erstes wird die Messreihe ohne Fenster geplottet
plt.figure(figsize=(15,10))
plt.plot(lamb_og, inten_og, label = 'ohne Fenster' )
#dann die mit Fenster
plt.plot(lamb_mg, inten_mg, label = 'mit Fenster' )
#unser Bild bekommt nocht einen Titel, Achsenbeschriftugn und eine Legende
plt.title( 'Gemessenes Sonnenspektrum mit und ohne Fenster' )
plt.xlabel( 'Wellenlaenge / nm' )
plt.ylabel( 'Intensitaet / b.E.' )
#plt.legend()
plt.grid() 
plt.ylim(( 0 , 30000 ))
plt.xlim(( 250 , 900))
#zum Schluss noch speichern
#plt.savefig("F:/WinPython-64bit-3.4.4.4Qt5/notebooks/data/AnnaLeoni/himmelohneglas.pdf", format="PDF")
```

```python
#Hier wird die Absorbption von Glas berechnet
#dabei werden die Werte, welche Null sind Ã¼bersprungen
A =1- inten_mg[inten_og>0] / inten_og[inten_og>0]
#Hier wird der zweite Plot initialisiert
plt.plot(lamb_mg[inten_og>0], A) 
plt.title( 'Absorption von Glas' ) 
plt.xlabel( 'Wellenlaenge / nm' ) 
plt.ylabel( 'Absorption / b.E.' ) 
#bei uns musste der Bereich auf negative Zahlen erweitert werden, da die Messwerte bei ohne Fenster teilweise kleiner sind, als die mit Fenster
plt.ylim(( 0 , 1 ))
plt.xlim(( 320 , 800 )) 
#datei speichern
plt.savefig( "F:/WinPython-64bit-3.4.4.4Qt5/notebooks/data/AnnaLeoni/absorptionglas.pdf" , format = "pdf" )
```
