```python
#import modules
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-white')
plt.rcParams['errorbar.capsize']=2
```

```python
#background
bg_time = 300
bg_prime = 118
bg_prime_err = np.sqrt(bg_prime)
bg = bg_prime/bg_time
bg_err = bg_prime_err/bg_time
```

```python
#absorption of beta-rays
#load data
layers_sr = np.arange(13)*0.3
layers_sr_err = 0.03*layers_sr

sr_prime = np.array([1248, 708, 498, 398, 252, 136, 89, 54, 167, 120, 76, 65, 50])
sr_prime_err = np.sqrt(sr_prime)
sr_time = np.array([30, 30, 30, 30, 30, 30, 30, 30, 120, 120, 120, 120, 120])
sr = sr_prime/sr_time - bg
sr_err = np.sqrt((sr_prime_err/sr_time)**2+bg_err**2)

#plot data
plt.figure('beta')
plt.yscale('log')
plt.ylabel('Zählrate [1/s]')
plt.xlim(0, 4.3)
plt.xlabel('Absorberdicke [mm]')
plt.title(r'Absortion von $\beta$-Strahlung')
plt.errorbar(layers_sr, sr, sr_err, layers_sr_err, linestyle = 'None')
plt.savefig('Diagramme/Sr.pdf')
```

```python
#absorption of gamma rays
#load data
layers_co = np.arange(11)*0.5
layers_co_err = layers_co*0.03

co_prime = np.array([2968, 2016, 1533, 1126, 843, 686, 487, 398, 298, 226, 168])
co_prime_err = np.sqrt(co_prime)
co_time = np.ones(11)*60
co = co_prime/co_time - bg
co_err = co_prime_err/co_time

#plot data
plt.figure('gamma')
plt.yscale('log')
plt.ylabel('Zählrate [1/s]')
plt.xlim(0, 6)
plt.xlabel('Absorberdicke [mm]')
plt.title(r'Absortion von $\gamma$-Strahlung')
plt.errorbar(layers_co, co, co_err, layers_co_err, linestyle = 'None')
plt.savefig('Diagramme/Co.pdf')
```

```python
#absorption of alpha-rays
#load data
p = np.array([27, 111, 204, 297, 403, 512, 602, 702, 820, 898, 983, 434, 462, 331, 381])
p_err = np.ones(15)

alpha_prime = np.array([4908, 4829, 4820, 4612, 346, 73, 82, 62, 76, 59, 79, 93, 78, 3783, 1005])
alpha_prime_err = np.sqrt(alpha_prime)
alpha_time = 60
alpha = alpha_prime/alpha_time
alpha_err = alpha_prime_err/alpha_time

#plot data
plt.figure('alpha')
plt.ylabel('Zählrate [1/s]')
plt.xlabel('Druck [mbar]')
plt.title(r'Absortion von $\alpha$-Strahlung')
plt.errorbar(p, alpha, alpha_err, p_err, linestyle = 'None')
plt.savefig('Diagramme/Alpha.pdf')
```
