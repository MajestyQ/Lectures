```python
import matplotlib.pyplot as plt
import numpy as np

f=np.array([3.13, 6.06, 9.09, 11.9, 15.1])
err_f=np.array([0.003, 0.006, 0.009, 0.012, 0.015])
U_pp=np.array([1.84, 4.32, 6.08, 8.4, 7.84])
err_U_pp=np.array([0.06, 0.13, 0.18, 0.25, 0.24])

plt.plot(f,U_pp, linestyle='none',marker='x', label='Messwert')
plt.errorbar(f, U_pp, xerr=err_f, yerr=err_U_pp)
plt.axis([3, 16, 1, 9])
plt.xlabel('Frequenz [Hz]')
plt.ylabel('Peak-Peak-Spannung [V]')
plt.title('Peak-Peak-Spannung in Abhängigkeit der Frequenz')

plt.show()
```
